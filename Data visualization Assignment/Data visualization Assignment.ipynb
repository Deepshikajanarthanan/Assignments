**In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart e=with transfromed dimensions and color each data point with specific class.**

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
x = PCA(n_components = 3).fit_transform(iris.data)
y = iris.target
```

```python
fig = plt.figure(1, figsize = (10,7))
ax = Axes3D(fig, elev = -150, azim = 120)


ax.scatter(x[:,0],x[:,1],x[:,2], c=y, cmap = plt.cm.prism)

ax.set_title("PCA Dimensions")
ax.set_xlabel("First dimension")
ax.set_ylabel("Second dimension")
ax.set_zlabel("Third dimension")
```
